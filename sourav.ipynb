```python
def check(ch):
 
    if (ch >= 'A' and ch <= 'Z'):
        print(ch,"is an UpperCase character")
    elif (ch >= 'a' and ch <= 'z'):
        print(ch,"is an LowerCase character")
    else:
        print(ch,"is a digit")

ch = input("Enter your responce : ")
check(ch)
```

```python
n = int(input("Enter your number: "))
s = n
p = len(str(n))
sum1 = 0
while n != 0:
    r = n % 10
    sum1 = sum1+(r**p)
    n = n//10
if s == sum1:
    print("The given number", s, "is armstrong number")
else:
    print("The given number", s, "is not armstrong number")
```

```python
from collections import Counter

def mean(data, n):
    data_sum = sum(data)
    result = data_sum/n
    print("The mean of the data is: " , result)

def median(data, n):
    data.sort()
    n = n//2
    # print(n)
    print("The median of the data is: " , data[n])

def mode(data):
    result = max(data, key = data.count)
    print("The mode of the data is: " , result)

data = [5,6,1,3,4,5,6,2,7,8,6,5,4,6,5,1,2,3,4]
n = len(data)

print("data = ", data)
print("""
Please select
1. Mean
2. Median
3. Mode
anything else will exit the code
""")
r = int(input())

if (r == 1):
    mean(data, n)
elif (r == 2):
    median(data, n)
elif (r == 3):
    mode(data)
else:
    print("Please select a suitable option")
```

```python
import matplotlib.pyplot as mp

month =['March','April','May','June','July','August']
Jeans = [1500,3500,6500,6700,6000,6800]
T_Shirts = [4400,4500,5500,6000,5600,6300]
Shirts = [6500,5000,5800,6300,6200,4500]

mp.plot(month, Jeans, label='Jeans', color='g', linestyle='dashed', linewidth=4,\
        marker='o', markerfacecolor='k', markeredgecolor='r')
mp.plot(month, T_Shirts, label='T_Shirts', color='b', linestyle='dashed', linewidth=4,\
        marker='3', markerfacecolor='k', markeredgecolor='g')
mp.plot(month, Shirts, label='Shirts', color='r', linestyle='dashed', linewidth=4,\
        marker='v', markerfacecolor='k', markeredgecolor='b')

mp.title("Clothes store data")
mp.xlabel("Months")
mp.ylabel("Sales")
mp.legend()
mp.show()
```
